```python
import pandas as pd
import numpy as np

psup_df = pd.read_csv("../outputs/psup.csv")
anom_df = pd.read_csv("../outputs/anomaly_score.csv")
```

```python
df = pd.concat([psup_df, anom_df], axis=1)
df.head()
```

```python
def normalize(x):
    return (x - x.min()) / (x.max() - x.min() + 1e-6)

df["psup_norm"] = normalize(df["psup"])
df["anomaly_norm"] = normalize(df["anomaly_score"])
```

```python
ALPHA = 0.4   # supervised weight
BETA  = 0.6   # unsupervised weight

df["final_risk"] = (
    ALPHA * df["psup_norm"] +
    BETA  * df["anomaly_norm"]
)
```
